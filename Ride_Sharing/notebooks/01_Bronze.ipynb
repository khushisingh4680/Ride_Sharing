```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
    .appName("Ride_Sharing_Bronze") \
    .getOrCreate()

print("Spark Session Created Successfully!")
```

```python
base_path = "/content/drive/MyDrive/Ride_Sharing"
```

```python
drivers = spark.read.csv(
    f"{base_path}/data/raw/drivers.csv",
    header=True,
    inferSchema=True
)

trips = spark.read.csv(
    f"{base_path}/data/raw/trips.csv",
    header=True,
    inferSchema=True
)

trip_logs = spark.read.csv(
    f"{base_path}/data/raw/trip_logs.csv",
    header=True,
    inferSchema=True
)
```

```python
print("========== Drivers Schema ==========")
drivers.printSchema()

print("========== Trips Schema ==========")
trips.printSchema()

print("========== Trip Logs Schema ==========")
trip_logs.printSchema()
```

```python
print("Drivers Data")
drivers.show(5)

print("Trips Data")
trips.show(5)

print("Trip Logs Data")
trip_logs.show(5)
```

```python
print("Number of Drivers :", drivers.count())
print("Number of Trips :", trips.count())
print("Number of Trip Logs :", trip_logs.count())
```

```python
drivers.write.mode("overwrite").parquet(
    f"{base_path}/data/bronze/drivers"
)

trips.write.mode("overwrite").parquet(
    f"{base_path}/data/bronze/trips"
)

trip_logs.write.mode("overwrite").parquet(
    f"{base_path}/data/bronze/trip_logs"
)

print("Bronze Layer Created Successfully!")
```

```python
bronze_drivers = spark.read.parquet(
    f"{base_path}/data/bronze/drivers"
)

bronze_trips = spark.read.parquet(
    f"{base_path}/data/bronze/trips"
)

bronze_logs = spark.read.parquet(
    f"{base_path}/data/bronze/trip_logs"
)

print("Bronze Drivers")
bronze_drivers.show(5)

print("Bronze Trips")
bronze_trips.show(5)

print("Bronze Trip Logs")
bronze_logs.show(5)
```

```python
spark.stop()
print("Spark Session Stopped Successfully!")
```
